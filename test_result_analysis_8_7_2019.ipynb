```python
import pandas as pd
df = pd.read_csv("test_result_8_7_2019", sep="\t")
```

```python
df = df.set_index("filename")
```

```python
df = df.drop(df.columns[0], axis=1)
df
```

```python
import matplotlib.pyplot as plt
plt.figure()
df_ss = pd.DataFrame()
df_ss["p_sen"] = df["p_or_not_sensitivity"]
df_ss["p_spe"] = df["p_or_not_specificity"]
df_ss["c_sen"] = df["c_or_not_sensitivity"]
df_ss["c_spe"] = df["c_or_not_specificity"]
df_ss.plot(kind='box')
plt.savefig("verify_new_data.png",dpi=200)
```
